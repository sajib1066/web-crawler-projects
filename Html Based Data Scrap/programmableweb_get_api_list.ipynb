# Python Web Crawler Project
## Get Api list from www.programmableweb.com

Import Modules

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Declare Url

```python
url = 'https://www.programmableweb.com/apis/directory'
```

Create Api Dictionary

```python
api_dict = {}
api_no = 0
```

Data Scrap Operation

```python
while True:
    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    api_list_1 = soup.find_all('tr', {'class': 'odd'})
    api_list_2 = soup.find_all('tr', {'class': 'even'})
    api_list = api_list_1 + api_list_2
    for api in api_list:
        link_tag = api.find('td', {'class': 'views-field views-field-title col-md-3'})
        name = link_tag.find('a').text
        link = 'https://www.programmableweb.com' + link_tag.find('a').get('href')
        category_data = api.find('td', {'class': 'views-field views-field-field-article-primary-category'})
        if category_data:
            category_tag = category_data.find('a')
        category = category_tag.text if category_tag else 'N/A'
        description_tag = api.find('td', {'class': 'views-field views-field-search-api-excerpt views-field-field-api-description hidden-xs visible-md visible-sm col-md-8'})
        description = description_tag.text if description_tag else 'N/A'
        api_no += 1
        api_dict[api_no] = [name, link, category, description]
    tag_url = soup.find('a', {'title': 'Go to next page'})
    if tag_url:
        if tag_url.get('href'):
            url = 'https://www.programmableweb.com' + tag_url.get('href')
            print(url)
        else:
            break
    else:
        break
print('Total Api: ', api_no)
```

Convert Dictionary To Pandas DataFrame

```python
api_dict_df = pd.DataFrame.from_dict(api_dict, orient='index', columns=['API Name', 'Link', 'Category', 'Description'])
```

Print DataFrame

Convert DataFrame To Csv file

```python
api_dict_df.to_csv('api_list.csv')
```
